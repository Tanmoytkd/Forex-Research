```python
import sqlalchemy as db2
engine = db2.create_engine('mysql://root@localhost/forex')
connection = engine.connect()
metadata = db2.MetaData()
renko = db2.Table('renko', metadata, autoload=True, autoload_with=engine)
state_table = db2.Table('state', metadata, autoload=True, autoload_with=engine)
price = db2.Table('price', metadata, autoload=True, autoload_with=engine)
# print(price.columns.keys())
```

```python
import sys
import pandas as pd

renkoData = connection.execute(db2.select([renko])).fetchall()
renkodf = pd.DataFrame(renkoData)
renkodf.columns = renkoData[0].keys()

row_count = len(renkodf.index)

decision_list = ['open']*row_count

renkodf['decision']=decision_list

level = 0

kangal_gushti = {
    'up': {},
    'down':{}
}

for index, row in renkodf.iterrows():
    if row['direction']=='up':
        level += 1
    if row['direction']=='down':
        level -= 1



    if (level+2) not in kangal_gushti['up']:
        kangal_gushti['up'][level+2]=[index]
    else:
        kangal_gushti['up'][level+2].append(index)
    
    if (level-2) not in kangal_gushti['down']:
        kangal_gushti['down'][level-2]=[index]
    else:
        kangal_gushti['down'][level-2].append(index)



    if level in kangal_gushti['up']:
        kangals = kangal_gushti['up'][level]
        for kangal in kangals:
            if renkodf.loc[kangal, :]['decision']=='open':
                ganja=1
                renkodf.at[kangal, 'decision']='up'
            kangal_gushti['up'][level].remove(kangal)
            
    if level in kangal_gushti['down']:
        kangals = kangal_gushti['down'][level]
        for kangal in kangals:
            if renkodf.loc[kangal, :]['decision']=='open':
                ganja=1
                renkodf.at[kangal, 'decision']='down'
            kangal_gushti['down'][level].remove(kangal)
            
renkodf.head(10)
```
